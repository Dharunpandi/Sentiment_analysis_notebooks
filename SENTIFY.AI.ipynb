```python
import pandas as pd
import numpy as np
%matplotlib inline

import spacy
nlp=spacy.load("en_core_web_sm")
```

```python
df=pd.read_csv("text.csv")
```

```python
df.head()
```

```python
df['sentiment']=df.label.map({
    0:"sadness",
    1:"joy",
    2:"love",
    3:"anger",
    4:"fear",
    5:"surprise"
})
```

```python
df.head()
```

```python
df.sentiment.value_counts()
```

```python
min_size=14972
```

```python
df_joy=df[df.sentiment=="joy"].sample(min_size,random_state=2022)
df_sadness=df[df.sentiment=="sadness"].sample(min_size,random_state=2022)
df_anger=df[df.sentiment=="anger"].sample(min_size,random_state=2022)
df_fear=df[df.sentiment=="fear"].sample(min_size,random_state=2022)
df_love=df[df.sentiment=="love"].sample(min_size,random_state=2022)
df_surprise=df[df.sentiment=="surprise"].sample(min_size,random_state=2022)
```

```python
df_balanced=pd.concat([df_joy,df_sadness,df_anger,df_fear,df_love,df_surprise])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df_balanced.text,df_balanced.label,test_size=0.2,random_state=2022,stratify=df_balanced.label)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
v=CountVectorizer()
```

```python
x_train_v=v.fit_transform(x_train)
```

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
```

```python
model.fit(x_train_v,y_train)
```

```python
x_test_v=v.transform(x_test.values)
```

```python
y_pred=model.predict(x_test_v)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_pred,y_test))
#testing performance
```

```python
def print_val(sentiment,arr):
    for i in range(len(arr)):
        if arr[i]==0:
            print(
                sentiment[i]," is about sadness")
        elif arr[i]==1:
            print(sentiment[i]," is about joy")
        elif arr[i]==2:
            print(sentiment[i]," is about love")
        elif arr[i]==3:
            print(sentiment[i]," is about anger")
        elif arr[i]==4:
            print(sentiment[i]," is about fear")
        else:
            print(sentiment[i]," is about surprise")
```
