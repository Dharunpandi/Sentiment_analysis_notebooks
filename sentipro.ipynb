```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn import *
import re
```

```python
df=pd.read_csv("sentimentdataset.csv")
```

```python
df.head()
```

```python
df=df.drop(columns=['Day','Hour','User','Platform','Retweets','Timestamp','Country','Likes','Month','Unnamed: 0.1','Unnamed: 0','Hashtags','Year'])
```

```python
df.head()
```

```python
def text_class(text):
    # Use re.sub() to replace non-alphabetic characters with an empty string
    text = re.sub(r'[\U00010000-\U0010ffff]', '', text)
    # Remove unwanted dots
    text=re.sub(r'!',"",text)
    return text
```

```python
df.Text=df.Text.apply(text_class)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
df.Sentiment=le.fit_transform(df.Sentiment)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(df.Text,df.Sentiment,shuffle=False, stratify=None, train_size=0.2)
```

```python
x_train
```

```python
df.head()
```

```python
vector=CountVectorizer()
```

```python
x_train_v=vector.transform(x_train)
```

```python
x_train_v.toarray()
```

```python
from sklearn import svm
model=svm.SVC(kernel='linear')
```

```python
model.fit(x_train_v,y_train)
```

```python
x_test_v=vector.fit_transform(x_test)
```

```python
model.predict(x_test_v)
```

```python
model.score(x_test_v,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
rf_classifier.fit(x_train_v,y_train)
```

```python
rf_classifier.score(x_test_v,y_test)
```
